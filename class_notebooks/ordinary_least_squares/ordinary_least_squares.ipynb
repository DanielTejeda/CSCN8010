```python
# generate data of 2 independent variables (X) and 1 dependent variable (y) with a linear relationship and a bit of noise

import numpy as np
import matplotlib.pyplot as plt

# generate 3 2D vectors, with elements each between 0 and 1
X = np.random.rand(3, 2)

y = 2*X[:,0] + 3*X[:,1] + np.random.rand(3)

# plot the data as a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], y, c='r', marker='o')
ax.set_xlabel('X0 Label')
ax.set_ylabel('X1 Label')
ax.set_zlabel('y Label')
plt.show()


```

```python
# generate 3 2D vectors, with elements each between 0 and 1
X = np.random.rand(100, 1)

y = 2*X[:,0] + np.random.rand(100)

# plot the data as a 3D scatter plot
fig = plt.figure()
#ax = fig.add_subplot(111, projection='3d')
ax = fig.add_subplot(111)
#ax.scatter(X[:,0], X[:,1], y, c='r', marker='o')
ax.scatter(X[:,0], y, c='r', marker='o')
ax.set_xlabel('X0 Label')
ax.set_ylabel('X1 Label')
#ax.set_zlabel('y Label')
plt.show()
```

```python
w = w = np.linalg.inv(X.T @ X) @ X.T @ y.reshape(-1,1)
```

```python
y_res = X @ w
y_res = y_res[:,0]
y_res
```

```python
# plot the data as a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(X[:,0], y_res, c='r')
ax.scatter(X[:,0], y)
ax.set_xlabel('X0 Label')
ax.set_ylabel('X1 Label')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X,y)

y_res2 = lr.predict(X)
y_res2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(X[:,0], y_res2, c='r')
ax.scatter(X[:,0], y)
ax.set_xlabel('X0 Label')
ax.set_ylabel('X1 Label')
plt.show()
```
